```python
import nbformat
import base64
import io
import os
from PIL import Image
import pytesseract

# Load the Jupyter Notebook
notebook_path = "1.1-basic-concepts.ipynb"

with open(notebook_path, "r", encoding="utf-8") as f:
    notebook = nbformat.read(f, as_version=4)

# Extract and process images
ocr_results = []

for cell in notebook.cells:
    if cell.cell_type == "markdown" and "attachments" in cell:
        for line in cell.source.split("\n"):
            if "![image" in line and "attachment:" in line:
                try:
                    # Extract base64 image key
                    img_key = line.split("attachment:")[1].split(")")[0]

                    # Check if the key exists in attachments
                    for img_format in ["image/png", "image/jpeg"]:
                        if img_key in cell["attachments"] and img_format in cell["attachments"][img_key]:
                            img_data = cell["attachments"][img_key][img_format]
                            img_bytes = base64.b64decode(img_data)

                            # OCR processing
                            image = Image.open(io.BytesIO(img_bytes))
                            extracted_text = pytesseract.image_to_string(image)

                            ocr_results.append(extracted_text)
                            break  # Process only one format (PNG/JPEG) per image
                except Exception as e:
                    print(f"Error processing image: {e}")

# Display extracted text
for idx, text in enumerate(ocr_results, start=1):
    print(f"Extracted Text {idx}:\n{text}\n{'-'*50}")

# Save OCR results to a file (optional)
with open("ocr_results.txt", "w", encoding="utf-8") as f:
    for idx, text in enumerate(ocr_results, start=1):
        f.write(f"Extracted Text {idx}:\n{text}\n{'-'*50}\n")
```
